## Cleopath Correlated Clustered Networks

## Basic single leaky neuron

```python
from matplotlib import pyplot as plt
import numpy as np
from scipy import special 
%matplotlib inline

time_decay = 0.01
time_arr = np.linspace(0, 10, 1000)

input_arr = 0
#input_arr[100:150] = 0

neuron_rate = np.zeros((time_arr.shape))
neuron_rate[0] = 0.9

for i in range(1,1000):
    neuron_rate[i] = neuron_rate[i-1] +  time_decay*(-neuron_rate[i-1]  + 
                                                     special.expit((input_arr + neuron_rate[i-1])*10))
```

```python
plt.plot(neuron_rate)
#plt.xlim((0, 10))
plt.show()
```

```python
values = np.linspace(-0.5, 1.5, 1000)
derivative = time_decay*(-values  + special.expit((input_arr + values)*4))
plt.plot(values, derivative)
#plt.ylim((-0.1, 1.1))
plt.show()
```

```python
x = np.linspace(-10,10,1000)
y = special.expit(x)
z = special.expit(x*10)
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

## Two recurrent neurons

```python
time_decay = 0.01
time_arr = np.linspace(0, 10, 1000)

NEURONS_N = 2

input_arr = -0.6
#input_arr[100:150] = 0

neuron_rate = np.zeros((NEURONS_N,time_arr.shape[0]))
neuron_rate[:,0] = np.array([0.9, 0])

weight_mtx = np.array([[0.5, -0.5],
              [-0.5, 0.5]])

input_vect = np.array(NEURONS_N)

for i in range(1,1000):
    neuron_rate[:,i] = neuron_rate[:,i-1] +  time_decay*(-neuron_rate[:,i-1]  + 
                                                     special.expit((weight_mtx.dot(neuron_rate[:,i-1]))*10))
```

```python
plt.plot(neuron_rate[0,:])
plt.plot(neuron_rate[1,:])
plt.show()
```
